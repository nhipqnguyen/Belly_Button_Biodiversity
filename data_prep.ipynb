```python
def color_code(s):
    c = ''
    if s == 'red':
        c = '#e01212'
    elif s == 'blue':
        c = '#0d48d4'
    elif s == 'green':
        c = '#41851c'
    elif s == 'emerald':
        c = '#5FFB17'
    elif s == 'light purple':
        c = '#a78cbd'
    elif s == 'navy':
        c = '#000080'
    elif s == 'black&pink':
        c = '#000000, #FFC0CB'
    elif s == 'black':
        c = '#000000'
    elif s == 'olive green':
        c = '#808000'
    elif s == 'purple':
        c = '#800080'
    elif s == 'white':
        c = '#FFFFFF'
    elif s == 'pink':
        c = '#FFC0CB'
    elif s == 'orange':
        c = '#FFA500'
    elif s == 'aquamarine':
        c = '#7FFFD4'
    elif s == 'yellow':
        c = '#FFFF00'
    return c
```

```python
df['color'] = [color_code(s) for s in df['What is your favorite color?']]
df
```

```python
names = []
for name in df['Full Name']:
    names.append(', '.join([name.split()[1],name.split()[0]]))
names
```

```python
df['last_first'] = names
df
```

```python
df.sort_values(by=['last_first'], inplace=True)
df
```

```python
df['id'] = range(1, len(df)+1)
df
```

```python
ids = df['id'].to_list()
metadata = []
for i in range(len(df)):
    metadata.append({'id': ids[i], 'name': df['last_first'][i],
                      'home town': df['Where is your hometown? (City, Country)'][i],
                     'pet': df['Do you have any pet(s)?'][i],
                     'morning/night':df['Are you a morning or night person?'][i],
                    'favorite color': df['What is your favorite color?'][i]})
metadata
```

```python
food = []
for i in range(len(df)):
    food.append({'id': ids[i], 
                 'fav_dishes': df['Please rank 5 of your favorite dishes.'][i].split(',')})
for f in food:
    f['fav_dishes'] = [t.strip().lower() for t in f['fav_dishes']]
food
```

```python
exercise = []
for i in range(len(df)):
    exercise.append({'id': ids[i], 'freq': df['How many times do you exercise per week?'][i]})
exercise
```

```python
survey = {'ids': ids, 'metadata': metadata, 'food': food, 'exercise': exercise}
with open('survey.json', 'w') as f:
    json.dump(survey, f)
```
